```python
class Estudiante:
    def __init__(self, nombre, apellido, cedula, email, semestre, materias):
        self._nombre = nombre
        self._apellido = apellido
        self.cedula = cedula
        self._email = email
        self.semestre = semestre
        self._materias = materias

    #Metodo de encapsulamiento

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, valor):
        self._nombre = valor

    @property
    def apellido(self):
        return self._apellido

    @apellido.setter
    def apellido(self, valor):
        self._apellido = valor

    @property
    def cedula(self):
        return self._cedula

    @cedula.setter
    def cedula(self, valor):
        if valor.isdigit() and len(valor) ==10:
            self._cedula = valor
        else:
            print("Cédula inválida: solo números y máximo 10 dígitos")

    @property
    def email(self):
        return self._email

    @email.setter
    def email(self, valor):
        self._email = valor

    @property
    def semestre(self):
        return self._semestre

    @semestre.setter
    def semestre(self, valor):
        if 1 <= valor <= 9:
            self._semestre = valor
        else:
            print("Semestre inválido: debe estar entre 1 y 9")

    @property
    def materias(self):
        return self._materias

    @materias.setter
    def materias(self, valor):
        self._materias = valor


    def __str__(self):
        return str(self.__dict__)

if __name__=="__main__":
    est = Estudiante("Ana", "Lopez", "1234567890", "ana@gmail.com", 3, ["Matemáticas", "Python"])
    print(est)

    # Modificar valores válidos
    est.semestre = 5
    print(est)

    # Probar validaciones
    est.cedula = "abc"
    est.semestre = 12
```
